```python
fiscal_years = [
    '2022-7888', '2021-7703', '2020-7500', '2019-6025',
    '2018-6021', '2017-6017', '2016-6013', '2015-6009', '2014-6005', 
    '2013-6001', '2012-5997', '2011-5993'
]

year_links = []

for year in fiscal_years:
    result_id = f'{year}-results'
    table = soup.find('div', attrs={'id': result_id}).find('a', attrs={'title': 'XBRL Spreadsheet'})
    
    if table:
        year_links.append(table['href'])  # Collect the link
    else:
        print(f"No table found for year {year}")
```
